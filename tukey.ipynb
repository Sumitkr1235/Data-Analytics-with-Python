```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_excel('Tensile strength of paper.xlsx')
df
```

```python
df.columns
```

```python
data_r1 = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['hardwood concentration 5%', 'hardwood concentration 10%',
       'hardwood concentration 15%', 'hardwood concentration 20%'])
data_r1.columns = ['index', 'treatments', 'value']
```

```python
data_r1
```

```python
model = ols('value ~ C(treatments)', data=data_r1).fit()
```

```python
model.summary()
```

```python
aov_table = sm.stats.anova_lm(model, typ=1)
aov_table
```
